```python
# stack and unstack -- used to change shape of dataframe
```

```python
ndf = df[(df.Edition == 2008)].groupby(['Gender','Sport','Medal']).agg({'Gender':['count']})
ndf
```

```python
df[(df.NOC == 'USA') & (df.Medal == 'Gold')].groupby(['NOC','Gender']).size().plot(kind='bar', x = 'Gender')
```

```python
df.groupby(['Athlete','Medal']).size().unstack('Medal',fill_value=0).sort_values(['Gold','Silver','Bronze'], ascending=False).head().plot(kind='bar')
```
